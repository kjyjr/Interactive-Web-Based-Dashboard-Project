```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
mongo = MongoClient(port=27017)
```

The 'museums' database and its collection, 'museums_us_orig,' were established in MongoDBCompass, and the collection populated
by importing a csv file, 'museums,' that was downloaded from Kaggle.com at: https://www.kaggle.com/datasets/imls/museum-directory. 

```python
print(mongo.list_database_names())
```

```python
db = mongo.museums
```

```python
print(db.list_collection_names())
```

```python
museums_us_orig = db.museums_us_orig
```

```python
museums_us_orig.count_documents({})
```

```python
museums_us_orig.find_one()
```

```python
museums_us_orig_df = pd.DataFrame(list(museums_us_orig.find()))
museums_us_orig_df.head()
# DataFrame code in this cell was sourced from: https://stackoverflow.com/questions/17805304/how-can-i-load-data-from-mongodb-collection-into-pandas-dataframe
```

```python
museums_us_orig_df.columns.to_list()
```

```python
museums_us_orig_df = museums_us_orig_df.loc[:, ['Museum ID', 'Museum Name', 'Museum Type', 
                    'Latitude','Longitude', 'Locale Code (NCES)','Income',
                         'Revenue','State (Administrative Location)']]
museums_us_orig_df.head()
```

```python
museums_us_orig_df = museums_us_orig_df.rename(columns= {
    'Museum ID': 'Museum_ID',
    'Museum Name': 'Museum_Name',
    'Museum Type': 'Museum_Type',
    'State (Administrative Location)': 'State',
    'Locale Code (NCES)': 'Locale_Code'
})
museums_us_orig_df.head()
```

```python
museums_us_orig_df.info()
```

```python
museums_us_new_df = museums_us_orig_df.dropna(how='any')
museums_us_new_df.info()
```

```python
museums_us_new_df = museums_us_new_df.sort_values("Revenue", ascending=False)
museums_us_new_df.head(3)
```

```python
museums_us_new_dict = museums_us_new_df.to_dict('records')
for r in range(3):
    pprint(museums_us_new_dict[r])
```

```python
museums_us_new = db.museums_us_new
museums_us_new.insert_many(museums_us_new_dict)
```

```python
museums_us_new.count_documents({})
```

```python
museums_us_new_results = museums_us_new.find()
for r in range(3):
    pprint(museums_us_new_results[r])
```
