```python
# Import the dependencies
import pandas as pd
from pprint import pprint
```

```python
museumDF = pd.read_csv("museums.csv")
museumDF.head()
```

```python
museumColumns = museumDF.columns.to_list()
museumColumns
```

```python
museumDF = museumDF.loc[:, ['Museum ID', 'Museum Name', 'Museum Type', 
                    'Latitude','Longitude', 'Locale Code (NCES)','Income',
                         'Revenue','State (Administrative Location)']]
museumDF.head()
```

```python
museumDF = museumDF.rename(columns= {
    'Museum ID': 'Museum_ID',
    'Museum Name': 'Museum_Name',
    'Museum Type': 'Museum_Type',
    'State (Administrative Location)': 'State',
    'Locale Code (NCES)': 'Locale_Code'
})
museumDF.head()
```

```python
museumDF = museumDF.dropna(how='any')
```

```python
museumDF = museumDF.sort_values("Revenue", ascending=False)
```

```python
museumDF['Museum_Type'].unique()
```

```python
museumDict = museumDF.to_dict('records')
pprint(museumDict)
```

```python
from pymongo import MongoClient
mongo = MongoClient(port=27017)
db =mongo.Project_3
museums = db.Museums_2
```

```python
museums.insert_many(museumDict)
```

```python
museumResult = museums.find()

for i in range(3):
    pprint(museumResult[i])
```
